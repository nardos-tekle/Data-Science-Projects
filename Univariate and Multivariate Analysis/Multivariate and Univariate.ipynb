```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
sns.scatterplot(x='carat', y='price', data= diamonds, color='#F15B2A') # Multivariate scatterplot on the variables price and carat 
```

```python
sns.jointplot(x='carat', y='price', data= diamonds, color='#F15B2A', kind='reg') # Multi variate joint plot demonstrating the relationship between the three variables (price, carat, and depth) 
```

```python

sns.catplot(x='cut', y='price', hue='depth', data=diamonds, kind='bar', errwidth=0)
```

```python
sns.displot(diamonds,color='#F15B2A', bins=50, kde=True) # Univariate distribution plots
            
```

```python
sns.kdeplot(diamonds) #Univariate kde plot
```
