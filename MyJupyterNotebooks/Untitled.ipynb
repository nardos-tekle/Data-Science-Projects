```python
house = data.houseAge # set a house variable to the houseAge column
```

```python
norm = (house - house.mean())/house.std() # under the norm variable normalised the houseAge column
```

```python
norm
```

```python
data.houseAge = norm # added the normalised data into the houseAge column
```

```python
data['houseAgeStandardized'] = norm # created a new column under the name houseAgeStandardized and added the normalised data
```
