```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
# importing libraries
```

```python
data = pd.read_csv('countries.csv')
data.head() # importing dataset
```

```python

# Filter data for Asian and African countries
asian_countries = data[data.continent == "Asia" ] 
african_countries = data[data.continent == "Africa"]

# Filter data for the year 1952
data_1952 = data[data.year == 1952]
```

```python
plt.hist(asian_countries.lifeExpectancy)
plt.title('Distribution of Life Expectancy in Asian Countries', fontsize=14)
plt.xlabel('Life Expectancy (years)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show
# displays the distribution of life expectancy in all Asian countries
```

```python
plt.hist(african_countries.lifeExpectancy)
plt.title('Distribution of Life Expectancy in African Countries', fontsize=14)
plt.xlabel('Life Expectancy (years)', labelpad= 20, fontsize=12)  
plt.ylabel('Frequency', fontsize=12)
plt.show # Displays the distribution of life expectancy in all African countries
```

```python
plt.scatter(data_1952.continent, data_1952.lifeExpectancy )
plt.title('Continents and Life Expectancy in 1952', fontsize=14)
plt.xlabel('Continents')  
plt.ylabel('Life Expectancy', fontsize=12)
plt.show # This consists of  all three variables of life expectancy in 1952 in all continents.  
```
